```python
# line chart

import matplotlib.pyplot as plt

year = [2000, 2005, 2010, 2015, 2020, 2025]
pop = [6.1, 6.5, 6.9, 7.3, 7.8, 8.1]

plt.plot(year, pop)
plt.show()
```

```python
# scatterplot

height = [165, 170, 175, 160, 180, 155, 190, 185, 172, 168]
weight = [65, 70, 75, 60, 85, 55, 95, 90, 68, 72]

plt.scatter(height, weight)
plt.show()
```

```python
# histogram

plt.hist(height)
plt.show()
```

```python
# titles

height = [165, 170, 175, 160, 180, 155, 190, 185, 172, 168]
weight = [65, 70, 75, 60, 85, 55, 95, 90, 68, 72]

plt.scatter(height, weight)

xlab = 'Height [cm]'
ylab = 'Weight [kg]'
title = 'Height vs Weight'

plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)

plt.show()
```

```python
# titles

height = [165, 170, 175, 160, 180, 155, 190, 185, 172, 168]
weight = [65, 70, 75, 60, 85, 55, 95, 90, 68, 72]

plt.scatter(height, weight)

tick_val = [100,160,220]
tick_lab = ['small', 'bigger', 'biggest']
plt.xticks(tick_val, tick_lab)

plt.show()
```

```python
# titles

height = [165, 170, 175, 160, 180, 155, 190, 185, 172, 168]
weight = [65, 70, 75, 60, 85, 55, 95, 90, 68, 72]
count = [30, 5, 40, 8, 2, 1, 9, 2, 40, 6]

plt.scatter(height, weight, s = count)

tick_val = [100,160,220]
tick_lab = ['small', 'bigger', 'biggest']
plt.xticks(tick_val, tick_lab)

plt.show()
```

```python
# colour

height = [165, 170, 175, 160, 180, 155, 190, 185, 172, 168]
weight = [65, 70, 75, 60, 85, 55, 95, 90, 68, 72]
count = [30, 5, 40, 8, 2, 1, 9, 2, 40, 6]

plt.scatter(height, weight, c = count, alpha = 0.8)

tick_val = [100,160,220]
tick_lab = ['small', 'bigger', 'biggest']
plt.xticks(tick_val, tick_lab)

plt.show()
```

```python
# annotations and grids

height = [165, 170, 175, 160, 180, 155, 190, 185, 172, 168]
weight = [65, 70, 75, 60, 85, 55, 95, 90, 68, 72]
count = [30, 5, 40, 8, 2, 1, 9, 2, 40, 6]

plt.scatter(height, weight, c = count, alpha = 0.8)

tick_val = [100,160,220]
tick_lab = ['small', 'bigger', 'biggest']
plt.xticks(tick_val, tick_lab)

plt.text(150, 70, 'test')
plt.text(190, 80, 'test2')

plt.grid(True)

plt.show()
```
